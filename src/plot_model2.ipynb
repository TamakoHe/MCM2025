```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'SimHei'
```

```python


# 设置随机种子以保证结果可复现（可选）
np.random.seed(42)

# 时间范围，假设从第 0 期到第 30 期，每期一个单位
time = np.linspace(0,31,1000)

# 人为划分两个阶段：
# 1) 前半段（破坏期）数量快速下降
# 2) 后半段（恢复期）数量逐渐上升
# 同时加入一定随机噪声模拟真实情况

population = []
for t in time:
    if t <= 10:
        # 简单示例：从 100 开始，每个时间点下降一点
        # 加入小范围噪声 np.random.normal(均值, 标准差)
        pop_val = 100 - 5 * t
    else:
        # 恢复期示例：使用 logistic 函数模拟逐渐回升
        # 中间 50 作为再次回升的基准值，0.3 决定上升速度，可根据需要调整
        pop_val = 50 + 50 / (1 + np.exp(-0.3 * (t - 10)))
        # 再加入少许噪声
        # pop_val += np.random.normal(0, 3)
    population.append(max(pop_val, 0))  # 保证数量不为负

# 绘图
plt.figure(figsize=(8, 5))
plt.plot(time, population, linestyle='-', color='blue')
plt.title("物种数量随时间变化示例（带波动）")
plt.xlabel("时间")
plt.ylabel("物种数量")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建数据
x = np.linspace(0, 10, 500)
y = np.sin(x)+2

# 创建图形和轴
fig, ax = plt.subplots()

# 绘制曲线
ax.plot(x, y, label='y = sin(x)')

# 添加x轴阴影部分
ax.fill_between(x, y, where=(x >= 2) & (x <= 6), color='gray', alpha=0.3, label='Shaded Area (2 <= x <= 6)')

# 添加标签和标题
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Curve with Shaded Area on x-axis')
ax.legend()

# 显示图形
plt.show()
```
