```python
import numpy as np
import matplotlib.pyplot as plt

# 定义降水量变化函数
def precipitation(m):
    """
    模拟东南亚降水量随月份的变化
    - 5~10月（雨季）降水量高
    - 11~4月（旱季）降水量低
    """
    return 0.5 * (1 - np.cos(2 * np.pi * (m - 2) / 12))

# 生成数据
months = np.arange(1, 13)  # 1~12月
rainfall = [precipitation(m) for m in months]

# 绘制曲线
plt.figure(figsize=(8, 4))
plt.plot(months, rainfall, marker='o', linestyle='-')
plt.xlabel("Month")
plt.ylabel("Relative Precipitation")
plt.title("Seasonal Precipitation Variation in Southeast Asia")
plt.xticks(months, [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])
plt.grid()
plt.show()
```
